```python
def plot_observables(L_values, T_vals, model_label="XY", logscale=False):
    fig, axs = plt.subplots(2, 2, figsize=(14, 10), constrained_layout=True)
    for L in L_values:
        if model_label.upper() == "XY":
            E = globals()[f'E_XY_{L}']
            M = globals()[f'M_XY_{L}']
            Cv = globals()[f'Cv_XY_{L}']
            Chi = globals()[f'Chi_XY_{L}']
        else:
            E = globals()[f'E_{L}']
            M = globals()[f'M_{L}']
            Cv = globals()[f'Cv_{L}']
            Chi = globals()[f'Chi_{L}']
            
        N = L**2 # Total number of spins for normalization
        
        # 1. Mean Energy per spin
        axs[0, 0].plot(T_vals, E / N, label=f'L = {L}')
        
        # 2. Mean Magnetization per spin
        axs[0, 1].plot(T_vals, abs(M) / N, label=f'L = {L}')
        
        # 3. Heat Capacity per spin
        axs[1, 0].plot(T_vals, Cv / N, label=f'L = {L}')
        
        # 4. Magnetic Susceptibility per spin
        axs[1, 1].plot(T_vals, Chi / N, label=f'L = {L}')
    # Format y-labels
    axs[0, 0].set_ylabel(r'Mean Energy per Spin $\langle E \rangle / L^2$')
    axs[0, 1].set_ylabel(r'Mean Magnetization per Spin $\langle |M| \rangle / L^2$')
    axs[1, 0].set_ylabel(r'Specific Heat $C_V / L^2$')
    axs[1, 1].set_ylabel(r'Magnetic Susceptibility $\chi / L^2$')
    
    #axs[1, 0].set_ylim(ymax=5,ymin=-1)
    #axs[1, 1].set_ylim(ymax=1000, ymin=-1)
    if logscale:
        axs[1, 0].set_yscale('log')
        axs[1, 1].set_yscale('log')
        
    # Format all subplots
    for ax in axs.flat:
        ax.set_xlabel('Temperature $T$')
        ax.grid(True, linestyle='--', alpha=0.6)
        ax.legend()
    fig.suptitle(f'{model_label} Model', fontsize=18)
    plt.show()
```

```python
def plot_spin_correlation(correlation_array, L, fit_interval, fit_type='exponential', degree=1):
    r = np.arange(len(correlation_array))
    c_r = np.array(correlation_array)
    
    
    # Filter data for the fitting interval
    r_min, r_max = fit_interval
    fit_mask = (r >= r_min) & (r <= r_max)
    r_fit = r[fit_mask]
    c_fit = c_r[fit_mask]
    
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Plot the full simulated data
    ax.scatter(r, c_r, label='Spin Correlation', color='royalblue', zorder=3)
    
    if fit_type == 'exponential':
        def exp_func(x, a, b):
            return a * np.exp(-b * x)
        
        p0 = (c_fit[0] if len(c_fit) > 0 else 1.0, 0.1)
        popt, _ = curve_fit(exp_func, r_fit, c_fit, p0=p0, maxfev=2000)
        ax.plot(r_fit, exp_func(r_fit, *popt), color='crimson', linewidth=2,
                label=f'Exp Fit: $a \cdot e^{{-r/\\xi}}$\n$\\xi = {1/popt[1]:.3f}$', zorder=4)
       
            
    elif fit_type == 'polynomial':
        coeffs = np.polyfit(r_fit, c_fit, degree)
        poly_func = np.poly1d(coeffs)
        ax.plot(r_fit, poly_func(r_fit), color='darkorange', linewidth=2,
                label=f'Poly Fit (deg={degree})', zorder=4)

    elif fit_type == 'power_law':
        def pow_func(x, a, b):
            return a * x**(-b)
        
        valid_mask = (r_fit > 0)
        p0 = (c_fit[valid_mask][0]*r_fit[valid_mask][0] if len(c_fit[valid_mask])>0 else 1.0, 0.25)
        popt, _ = curve_fit(pow_func, r_fit[valid_mask], c_fit[valid_mask], p0=p0, maxfev=2000)
        ax.plot(r_fit[valid_mask], pow_func(r_fit[valid_mask], *popt), color='red', linewidth=2,
            label=f'Power-law Fit: $a \cdot r^{{-\\eta}}$\n$\\eta = {popt[1]:.3f}$', zorder=4)

    else:
        print(f"Unknown fit_type: {fit_type}")

    # Formatting rules
    ax.set_xlabel('Distance $r$', fontsize=12)
    ax.set_ylabel('Spin Correlation $C(r)$', fontsize=12)
    ax.set_title(f'Spin Correlation Function ($L={L}$)', fontsize=14)
    
    # ax.set_yscale('log')
    #if fit_type == 'exponential':
    #     ax.set_xscale('log') 
    
    ax.legend()
    
    plt.tight_layout()
    plt.show()
```
