```python
def plot_observables(L_values, T_vals, model_label="XY", logscale=False):
    fig, axs = plt.subplots(2, 2, figsize=(14, 10), constrained_layout=True)
    for L in L_values:
        if model_label.upper() == "XY":
            E = globals()[f'E_XY_{L}']
            M = globals()[f'M_XY_{L}']
            Cv = globals()[f'Cv_XY_{L}']
            Chi = globals()[f'Chi_XY_{L}']
        else:
            E = globals()[f'E_{L}']
            M = globals()[f'M_{L}']
            Cv = globals()[f'Cv_{L}']
            Chi = globals()[f'Chi_{L}']
            
        N = L**2 # Total number of spins for normalization
        
        # 1. Mean Energy per spin
        axs[0, 0].plot(T_vals, E / N, label=f'L = {L}')
        
        # 2. Mean Magnetization per spin
        axs[0, 1].plot(T_vals, abs(M) / N, label=f'L = {L}')
        
        # 3. Heat Capacity per spin
        axs[1, 0].plot(T_vals, Cv / N, label=f'L = {L}')
        
        # 4. Magnetic Susceptibility per spin
        axs[1, 1].plot(T_vals, Chi / N, label=f'L = {L}')
    # Format y-labels
    axs[0, 0].set_ylabel(r'Mean Energy per Spin $\langle E \rangle / L^2$')
    axs[0, 1].set_ylabel(r'Mean Magnetization per Spin $\langle |M| \rangle / L^2$')
    axs[1, 0].set_ylabel(r'Specific Heat $C_V / L^2$')
    axs[1, 1].set_ylabel(r'Magnetic Susceptibility $\chi / L^2$')
    
    #axs[1, 0].set_ylim(ymax=5,ymin=-1)
    #axs[1, 1].set_ylim(ymax=1000, ymin=-1)
    if logscale:
        axs[1, 0].set_yscale('log')
        axs[1, 1].set_yscale('log')
        
    # Format all subplots
    for ax in axs.flat:
        ax.set_xlabel('Temperature $T$')
        ax.grid(True, linestyle='--', alpha=0.6)
        ax.legend()
    fig.suptitle(f'{model_label} Model', fontsize=18)
    plt.show()
```
